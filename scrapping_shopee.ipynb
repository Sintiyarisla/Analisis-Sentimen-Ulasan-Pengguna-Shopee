```python
from google_play_scraper import app, reviews, Sort

scrapreview,_ = reviews(
    'com.shopee.id',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=10000             # Jumlah maksimum ulasan yang ingin diambil
)
```

```python
import pandas as pd

df = pd.DataFrame(scrapreview)

df.to_csv("scrap_shopee.csv", index=False, encoding="utf-8-sig")
print(f"Total review yang disimpan: {len(df):,}\ndisimpan dalam 'scrap_shopee.csv'")
```
