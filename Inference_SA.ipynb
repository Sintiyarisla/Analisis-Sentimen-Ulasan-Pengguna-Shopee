# XGBoost

```python
import joblib

# Load model bundle
bundle = joblib.load('sentiment_model_XGBoost_TFIDF_70_30.joblib')

model = bundle['model']
tfidf = bundle['tfidf']
label_map = bundle['label_map']

# Data uji
texts = [
    "produk ini sangat bagus dan memuaskan",
    "pelayanan buruk dan mengecewakan"
]

# Transform
X_test_tfidf = tfidf.transform(texts)

# Predict
pred = model.predict(X_test_tfidf)
pred_label = [label_map[i] for i in pred]

print(pred_label)
```

```python
from tensorflow.keras.models import load_model
import joblib
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences


# Load model & config
lstm_model = load_model("sentiment_model_LSTM_W2V_80_20.h5")
config = joblib.load("sentiment_model_LSTM_W2V_80_20_config.joblib")

tokenizer = config['tokenizer']
label_map = config['label_map']
max_len = config['max_len']

# Contoh data uji
texts = [
    "produk ini sangat bagus dan memuaskan",
    "pelayanan sangat buruk dan mengecewakan"
]

# Preprocessing
seq = tokenizer.texts_to_sequences(texts)
pad = pad_sequences(seq, maxlen=max_len)

# Predict
pred_prob = lstm_model.predict(pad)
pred_enc = np.argmax(pred_prob, axis=1)
pred_label = [label_map[i] for i in pred_enc]

print(pred_label)
```

# GRU

```python
from tensorflow.keras.models import load_model
import joblib
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load model & config
gru_model = load_model("sentiment_model_GRU_W2V_80_20.h5")
config = joblib.load("sentiment_model_GRU_W2V_80_20_config.joblib")

tokenizer = config['tokenizer']
label_map = config['label_map']
max_len = config['max_len']

# Contoh teks uji
texts = [
    "aplikasinya sangat membantu dan mudah digunakan",
    "fiturnya jelek banget"
]

# Preprocessing
seq = tokenizer.texts_to_sequences(texts)
pad = pad_sequences(seq, maxlen=max_len)

# Predict
pred_prob = gru_model.predict(pad)
pred_enc = np.argmax(pred_prob, axis=1)
pred_label = [label_map[i] for i in pred_enc]

print(pred_label)
```

# BiLSTM

```python
from tensorflow.keras.models import load_model
import joblib
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load model & config
bilstm_model = load_model("sentiment_model_BiLSTM_80_20.h5")
config = joblib.load("sentiment_model_BiLSTM_80_20_config.joblib")

tokenizer = config['tokenizer']
label_map = config['label_map']
max_len = config['max_len']

# Contoh input teks
texts = [
    "produknya sangat bagus dan pengirimannya cepat",
    "aplikasinya sering error dan mengecewakan"
]

# Preprocessing
seq = tokenizer.texts_to_sequences(texts)
pad = pad_sequences(seq, maxlen=max_len)

# Predict
pred_prob = bilstm_model.predict(pad)
pred_enc = np.argmax(pred_prob, axis=1)
pred_label = [label_map[i] for i in pred_enc]

print(pred_label)
```
